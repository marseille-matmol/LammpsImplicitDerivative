## Strain matrix
Idea is to apply effective 2D strain:
$
\begin{equation}
    \left[
    \begin{matrix}
    \sigma_{xx} \\
    \sigma_{yy} \\
    \sigma_{zz} \\
    \sigma_{xy} \\
    \sigma_{xz} \\
    \sigma_{yz}
    \end{matrix}
    \right]
    =
    \left[
    \begin{matrix}
    {\rm C}_{11} & {\rm C}_{12} & {\rm C}_{12} & 0& 0& 0\\
    {\rm C}_{12} & {\rm C}_{11} & {\rm C}_{12} & 0& 0& 0\\
    {\rm C}_{12} & {\rm C}_{12} & {\rm C}_{11} & 0& 0& 0\\
    0 & 0 & 0 & {\rm C}_{44} & 0 & 0  \\
    0 & 0 & 0 & 0 & {\rm C}_{44} & 0 \\
    0 & 0 & 0 & 0 & 0 & {\rm C}_{44}
    \end{matrix}
    \right]
    \left[
    \begin{matrix}
    \epsilon_{xx} \\
    \epsilon_{yy} \\
    \epsilon_{zz} \\
    \epsilon_{xy} \\
    \epsilon_{xz} \\
    \epsilon_{yz}
    \end{matrix}
    \right]
\end{equation}
$

$$
E = \frac{V_0}{2}
\left[
    \begin{matrix}
    \epsilon_{xx} & \epsilon_{xy} & \epsilon_{xz} \\ 
    \epsilon_{xy} & \epsilon_{yy} & \epsilon_{yz} \\ 
    \epsilon_{xz} & \epsilon_{yz} & \epsilon_{zz}
    \end{matrix}
\right]
:
\left[
    \begin{matrix}
    \sigma_{xx} & \sigma_{xy} & \sigma_{xz} \\ 
    \sigma_{xy} & \sigma_{yy} & \sigma_{yz} \\ 
    \sigma_{xz} & \sigma_{yz} & \sigma_{zz}
    \end{matrix}
\right]
$$

$$
A = \frac{{\rm C}_{11} - {\rm C}_{12}}{2{\rm C}_{44}}
$$

```python
J_to_eV = 1.0/1.602176634e-19
m_to_A = 1.0e10
GPa_to_eVA3 = 1.0e9 * J_to_eV / m_to_A**3

# for W
C11_C12_C44 = np.array([509.,200.,144.]) * GPa_to_eVA3

def elastic_energy_density(strain):
    # for cubic materials
    C11_C12 = np.ones((3,3)) * C11_C12_C44[1]
    C11_C12 += np.diag(np.ones(3) * (C11_C12_C44[0] - C11_C12_C44[1]))
    
    stress = np.zeros((3,3))
    for i in range(3):
        for j in range(3):
            if i!=j:
                stress[i][j] = C11_C12_C44[2] * strain[i][j]
            else:
                stress[i][j] = (C11_C12 @ np.diag(strain))[i]
    return 0.5 * (stress*strain).sum()
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,5),sharey=True)
for i,ax in enumerate(axs):
    N = 4 + i
    start_script = f"""
                region C block 0 {N} 0 {N} 0 {N} units lattice
                create_box 2 C
                create_atoms 2 region C
                
                """
    L = SimpleLammps(	potential=WX_potential_data,\
                        start_script=start_script
                    )

    C0 = L.cell.copy()
    V0 = np.linalg.det(C0)
    epsilons = np.linspace(-0.03,0.03,11)
    E = []
    E_el = []

    for epsilon in epsilons:
        M = np.diag([epsilon,epsilon,-2.0*epsilon])
        
        L.apply_strain(M)
        E += [L.get_E()]
        E_el += [elastic_energy_density(M) * V0 ]
        L.apply_strain(np.linalg.inv(M+np.eye(3))-np.eye(3))

    E = np.array(E) / N**3
    E_el = np.array(E_el) / N**3
    E -= E.min()

    ax.plot(100*epsilons,E,'o-',label='MD')
    ax.plot(100*epsilons,E_el,'o-',label='Elastic')
    ax.set_xlabel('Strain Magnitude (%)')
    ax.legend()
```

We define a strained energy from some lattice ${\rm a}_0$ as $E({\rm a}_0,\boldsymbol{\epsilon})$,
where $\boldsymbol{\epsilon}=\boldsymbol{\epsilon}^\top\in\mathbb{R}^{3\times3}$ 

The lattice constant is defined as 
$$
\boldsymbol{\epsilon}=\epsilon\mathbb{I}_3
,\quad{\rm a}^* = (1+\epsilon^*){\rm a}_0
,\quad
\epsilon^* = \arg\min_\epsilon E({\rm a}_0,\epsilon\mathbb{I}_3)
$$

Linear potential energy for the same lattice:
$$
E({\rm a}_0,\boldsymbol{\epsilon}) = \boldsymbol{\Theta}^\top{\bf D}({\rm a}_0,\boldsymbol{\epsilon})
$$

$$
\boldsymbol{\epsilon}=\epsilon\mathbb{I}_3
,\quad
{\rm a}^* = (1+\epsilon^*){\rm a}_0
,\quad

\frac{\partial E}{\partial \epsilon} \Big|_{\epsilon=\epsilon^*}
=
\boldsymbol{\Theta}^\top\frac{\partial{\bf D}}{\partial \epsilon}\Big|_{\epsilon=\epsilon^*}
=0
$$

We want to make functions like

$$
{\rm a}^*({\boldsymbol{\Theta}}),\quad
{\rm C}_{11}({\boldsymbol{\Theta}}),\quad
{\rm C}_{12}({\boldsymbol{\Theta}}),\quad
{\rm C}_{44}({\boldsymbol{\Theta}})
$$

$$
{\bf C}\boldsymbol{\epsilon}=\frac{\partial E}{\partial\boldsymbol{\epsilon}}
=
\boldsymbol{\Theta}^\top\frac{\partial{\bf D}}{\partial\boldsymbol{\epsilon}}
$$
$$
\frac{1}{2}{\rm Tr}(\boldsymbol{\epsilon}{\bf C}\boldsymbol{\epsilon})=
\boldsymbol{\Theta}^\top{\bf D}(\boldsymbol{\epsilon})
$$


$$
{\bf A} = \bigoplus_{r=0}^R {\bf D}(\boldsymbol{\epsilon}_r)\in \mathbb{R}^{N_D\times R}
$$


Then we can have loss functions like 
$$
\mathcal{L}(\boldsymbol{\Theta}) = \|{\bf X}_0-{\bf X}^*({\boldsymbol{\Theta}}) + \lambda\|a^*-a^*({\boldsymbol{\Theta}})\|^2
$$
